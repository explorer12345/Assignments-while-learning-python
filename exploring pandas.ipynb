Pie chart distribution of different regions 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('sslc1.csv',sep=';')
#code for plotting pie plot
dfcounts=df['region'].value_counts()
dfcounts.plot.pie(figsize=(5,5),title='Regionwise distribution',autopct='%.2f')
```

Box plots

```python
from ipywidgets import interact
def f(Region):
    dfA=df[df.region==Region]
    del dfA['roll_number'],dfA['name'],dfA['total'],dfA['pass'],dfA['withheld'],dfA['extra'],dfA['region']
    dfA=dfA.replace('AA',np.nan)
    dfA['sl']=pd.to_numeric(dfA['sl'])
    dfA['fl']=pd.to_numeric(dfA['fl'])
    dfA['math']=pd.to_numeric(dfA['math'])
    dfA['sci']=pd.to_numeric(dfA['sci'])
    dfA['ss']=pd.to_numeric(dfA['ss'])
    dfA.plot.box()
    print(Region)
    plt.show()
    return
interact(f,Region={'A':'A','B':'B','C':'C','D':'D','E':'E','All':'A'and'B'and'C'and'D'and'E'});
```

```python
from ipywidgets import interact
def g(Region):
    dfp=df[df.region==Region]
    dfp=dfp.fillna('Fail')
    dfp=dfp.replace('P','Pass')
    dfpcounts=dfp['pass'].value_counts()
    dfpcounts.plot.pie(figsize=(5,5),title='Regionspecific pass percentage',autopct='%.2f')
    print(Region)
    plt.show()
    return
interact(g,Region=['A','B','C','D','E']);
```

Finding C.I. using Normal Distribution

```python
import ipywidgets as widgets
dfA=df.replace('AA',np.nan)
del dfA['roll_number'],dfA['name'],dfA['total'],dfA['pass'],dfA['withheld'],dfA['extra']
dfA=dfA[dfA.region=='A']
dfA['sl']=pd.to_numeric(dfA['sl'])
dfA['fl']=pd.to_numeric(dfA['fl'])
dfA['math']=pd.to_numeric(dfA['math'])
dfA['sci']=pd.to_numeric(dfA['sci'])
dfA['ss']=pd.to_numeric(dfA['ss'])
def h(n):#n=sample size
    dfs=dfA['math'].sample(n)
    x=dfs.mean()
    sigma=dfs.std()
    lower_limit=x-1.96*(sigma/n**0.5)
    upper_limit=x+1.96*(sigma/n**0.5)
    print('Confidence interval is [',end=" ")
    print(lower_limit,end=" ")
    print(',',end=" ")
    print(upper_limit,end=" ")
    print(']')
    return
interact(h, n=widgets.IntSlider(min=1,max=36401,step=1,value=10));#n=sample size
```

Finding C.I. using sampling

```python
def i(n):#n=sample size of each sample
    mean=[]
    for j in range(1000):
        dfa=dfA['math'].sample(n,replace=True)
        mean.append(dfa.mean())
    Mean=pd.DataFrame({'index':range(1000),'mu':mean},columns=['index','mu'])
    lower_limit=Mean['mu'].quantile(0.025)#2.5 percentile value
    upper_limit=Mean['mu'].quantile(0.975)#97.5 percentile value
    print('Confidence interval is [',end=" ")
    print(lower_limit,end=" ")
    print(',',end=" ")
    print(upper_limit,end=" ")
    print(']')
    return 
interact(i, n=widgets.IntSlider(min=1,max=36401,step=1,value=10));
```

Thus the values obtained from both methods are quite close
